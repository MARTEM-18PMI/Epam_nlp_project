```python
import string
from unidecode import unidecode
import nltk
from pymystem3 import Mystem
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
import pandas as pd
```

```python
def convert_numbers(string):
    num_dict = {'0' : "ноль",
                '1' : "один",
                '2' : "два",
                '3' : "три",
                '4' : "четыре",
                '5' : "пять",
                '6' : "шесть",
                '7' : "семь",
                '8' : "восемь",
                '9' : "девять"}
    for key, value in num_dict.items():
        string = string.replace(key, value)
    return string
```

```python
def remove_punctuation_and_stopwords(s):
    s = s.lower()
    
    punct = set(string.punctuation)
    # add some symbols to delete
    punct.add('»')
    punct.add(' »')
    punct.add(' «')
    punct.add('—')
    punct.add('«')
    punct.add('–')
    punct.add('№')
    
    for item in punct:
        s = s.replace(item, '')
    stop_words = set(stopwords.words('russian'))
    
    # split 
    word_tokens = word_tokenize(s)
    return (" ".join([item for item in word_tokens if not item in stop_words])).strip()
```

```python
def split_(s):
    return [i for i in s]

def lemma(text):
    rus_alph = [chr(i) for i in range(ord('а'),ord('а')+32)] + [chr(i) for i in range(ord('А'),ord('А')+32)]
    stem = Mystem()
    merged_text = '♥'.join(text)
    #print(len(text))
    doc = []
    res = []
    for t in stem.lemmatize(merged_text):
        if t != '♥':
            if (t != ' ' and t != '\n' and t != '\u200e'):
                for letter in split_(t):
                    if letter in rus_alph:
                        continue
                    else:
                        t = unidecode(t)
                doc.append(t)
        else:
            res.append(doc)
            doc = []
    res.append(doc)
    return res
```

```python
def convert(text):
    prepared = []
    for item in text:
        temp = remove_punctuation_and_stopwords(item)
        temp = convert_numbers(temp)
        prepared.append(temp)
    lemma_data = lemma(prepared)
    #print(len(lemma_data))
    result = []
    for item in lemma_data:
        result.append("".join(item))
    return result
```

```python
data = pd.read_excel("articles_1.xls")
data = data.drop(['id'], axis = 1)
data
```

```python
len(data)
```

```python
part_size = len(data)

titles = []
texts = []

for index, row in data.iterrows():
    titles.append(row[0])
    texts.append(row[1])

# list of titles
title_part = [titles[i: i+part_size] for i in range(0, len(titles), part_size)]
title_part = title_part[0]

# list of article texts
text_part = [texts[i: i+part_size] for i in range(0, len(texts), part_size)]
text_part = text_part[0]
```

```python
(len(title_part), len(text_part))
```

```python
res_titles = convert(title_part)
res_texts = convert(text_part)
```

```python
(len(res_titles), len(res_texts))
```

```python
dd = pd.DataFrame({'title' : res_titles, 'texts' : res_texts})
dd['date'] = data['date']
dd['tag'] = data['tag']
dd
```
