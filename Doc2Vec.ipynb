```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import pandas as pd
import numpy as np
from nltk.tokenize import word_tokenize
```

```python
data = pd.read_excel("articles_preprocess.xls")
data.drop('id', axis = 1, inplace = True)
data
```

```python
titles = [item[1][0] for item in data.iterrows()]
texts = [item[1][1] for item in data.iterrows()]
```

```python
tagged_titles = [TaggedDocument(word_tokenize(title), [i]) for i, title in enumerate(titles)]
tagged_texts = [TaggedDocument(word_tokenize(text), [i]) for i, text in enumerate(texts)]
```

```python
vector_size = 300
min_count = 1
epochs = 20
workers = 4

doc2vec = Doc2Vec(vector_size = vector_size, min_count = min_count,epochs = epochs, workers = workers)
doc2vec.build_vocab(tagged_titles + tagged_texts)
#doc2vec.build_vocab(tagged_texts, update=True)
```

```python
doc2vec.train(tagged_titles + tagged_texts, total_examples=doc2vec.corpus_count, epochs = doc2vec.iter)
```

```python
vectors_of_titles = []
vectors_of_texts = []

for i in range(data.shape[0]):
    vectors_of_titles.append(doc2vec.infer_vector(word_tokenize(titles[i])))
    vectors_of_texts.append(doc2vec.infer_vector(word_tokenize(texts[i])))
```

```python
data['titles_vec'] = vectors_of_titles
data['texts_vec'] = vectors_of_texts
data
```
